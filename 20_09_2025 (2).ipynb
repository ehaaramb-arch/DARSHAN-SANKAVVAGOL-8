```python
class BCA_1St_Semester :
    def student(self,a,b=0,c=0,d=0):
      return a+b+c+d
BCA = BCA_1St_Semester()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,0,4,-12))

```

```python
class Animal:
    def sound(self):
      print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
       print("Bow BOW Bow")


class Cat  (Animal):
    def sound(self):
       print("Meow MeOW Meow")

class Snake (Animal) :
   def sound(self):
    print("Buussssssssss")

class Cow (Animal) :
   def sound(self):
    print('Ammbbbbbbbbaaaa')

obj1 = Animal()
obj2 = Dog()
obj3 = Cat()
obj4 = Snake()
obj5 = Cow()

obj1.sound()
obj2.sound()
obj3.sound()
obj4.sound()
obj5.sound()
```
